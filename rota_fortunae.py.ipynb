```python
# PROBLEM 5 ROTA FORTUNAE
if __name__ == '__main__':
    secretWord = str(input("What is the word you want the players to guess? "))
    print(secretWord) # to be removed
    
    # array to store each letter of secret word, hidden from player
    hiddenWord = []
    blanks = []
    count = 0
    prevGuesses = []
    x = False
    win = False
    # prints an underscore for each character in the secret word
    for i in secretWord :
        hiddenWord.append(i)
        blanks.append("-")
        #print("_", end ="")
    print("\n") # prints blank line
    # print(hiddenWord) # to be hidden
    
    # displays _'s to player
    for j in blanks :
        print(j, end="")
  
    while count != len(secretWord) and win != True:
        print("\n") # prints blank line
        # prompts user for guess
        guess = str(input("Guess a letter, or \"solve\":"))
        if guess == "solve": #check if you want to solve
                solve_input = str(input("What is the entire puzzle?"))
                if solve_input == secretWord:
                    x= True
                    print("you are right")
                    win = True
                elif solve_input != secretWord:
                    print("no you're wrong")
                    x = True #so it wont do the solve again
        else:# checking if user's guess matches letter in secret word
            guess = guess[0]
        # print("user\'s guess:", guess) # to be removed
            for i in range(len(secretWord)) :
                if guess == secretWord[i]: # if match, print character
                    blanks[i] = guess
                    x = True
                    if (guess not in prevGuesses) :
                        prevGuesses.append(guess)
                        count += 1
                    elif (guess in prevGuesses):
                        print("You have guessed that letter before")
                    #print(blanks)

                    for j in blanks :
                        print(j, end="")

                
                 
        if(x == False): # if guess not in secretWord
            if (guess not in prevGuesses) : # if not already guessed
                print("There are no " + guess + " in the word") 
                prevGuesses.append(guess)
            else :
                print("You have guessed this letter before, but it is still not in the word")
                
            
            
       
    print("\n\nYou guessed all the letters !")  
```
