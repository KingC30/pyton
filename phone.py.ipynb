```python
"""
File:         phone.py
Author:       Gia Bao Vo
Date:         11/30/2020
Description:  This program make the phones number and the adress work

"""

"""
    Phone Class Starter Code

    This code defines the basic functionality that you need from a phone.
    When these functions are called they should communicate with the
    switchboards to find a path
"""
class Phone:
    def __init__(self, number, switchboard):
        """
        :param number: the phone number without area code
        :param switchboard: the switchboard to which the number is attached.
        """
        self.number = number
        self.switchboard = switchboard
        self.connected_to = None
        self.connect_str = None
        # you will need more parameters/attributes

    def connect(self, area_code, other_phone_number):
        """
        :param area_code: the area code of the other phone number
        :param other_phone_number: the other phone number without the area code
        :return: this you must decide in your implementation
        """
        phone_path = self.switchboard.connect_call(area_code, other_phone_number, [])
        for phone_object in phone_path.phones:
            if other_phone_number == phone_object.number:
                other_phone = phone_object

        self_num = str(self.switchboard.area_code) + "-" + str(self.number)
        dest_num = str(area_code) + "-" + str(other_phone_number)

        #if can connect
        if (self.connected_to == None):
            self.connect_str = dest_num

            #connect on dest phone as well
            self.connected_to = other_phone
            other_phone.connected_to = self
            other_phone.connect_str = self.number
            self.connected_to.connect_str = self_num

            self.is_connected = True
            print(self_num, "and", dest_num, "are now connected.")
        #else cannot connect
        else:
            print(self_num, "and", dest_num, "were not connected.")
        pass

    def disconnect(self, area_code, other_phone_number):
        """
        This function should return the connection status to disconnected.  You need
        to use new members of this class to determine how the phone is connected to
        the other phone.

        You should also make sure to disconnect the other phone on the other end of the line.
        :return: depends on your implementation
        """

        self_num = str(self.switchboard.area_code) + "-" + str(self.number)
        dest_num = str(area_code) + "-" + str(other_phone_number)

        #if can disconnect
        if (self.connected_to != None):
            self.disconnect_str = dest_num

            #disconnect on dest phone as well
            self.connected_to.connected_to = None
            self.connected_to = None

            print(self_num, "and", dest_num, "are now disconnected.")
        #else cannot disconnect
        else:
            print(self_num, "and", dest_num, "were not disconnected.")

        pass
```
